```python
labels = ["Without rewards", "With rewards"]
values = [1041, 1486]

plt.figure(figsize=(8,5))
bars = plt.bar(labels, values, color=["#999999","#4CAF50"])

# Annotate bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 30,
             f"${height:,}", ha='center', va='bottom', fontsize=12)

# Highlight % increase text
plt.text(0.5, max(values) - 200, "+43%", ha="center", fontsize=14, color="darkgreen", weight="bold")

plt.title("Tiltify Performance: Fundraising with vs without Rewards")
plt.ylabel("Average Raised ($)")
plt.ylim(0, 1700)
plt.tight_layout()
```
